```python
#import libraries
import pandas as pd
import requests
import os
```

```python
#make directories
if os.path.exists:
  print("Already exists")
else:
  os.makedirs("data/raw/tlc", exist_ok=True)
  os.makedirs("data/raw/lookup", exist_ok=True)
```

```python
#download parquet and csv files
parquet_url = 'https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-01.parquet'
csv_url = 'https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv'

raw_yellow_trip_data = requests.get(parquet_url)
raw_taxi_zone_lookup = requests.get(csv_url)

#file path directory locations
parquet_path = "data/raw/tlc/yellow_tripdata_2023-01.parquet"
csv_path = "data/raw/lookup/taxi_zone_lookup.csv"

#assign location to files
with open (parquet_path, "wb") as f:
  f.write(raw_yellow_trip_data.content)

with open (csv_path, "wb") as f:
  f.write(raw_taxi_zone_lookup.content)
```

```python
#load datasets
yellow_tripdata_parquet = pd.read_parquet(parquet_path)
taxi_zone_lookup_csv = pd.read_csv(csv_path)
```

```python
#assign datasets as dataframes using pandas
yellow_tripdata_df = pd.DataFrame(yellow_tripdata_parquet)
taxi_zone_lookup_df = pd.DataFrame(taxi_zone_lookup_csv)
```

```python
#Yellow Trip Data
print("Number of Rows: ")
print(yellow_tripdata_df.shape[0])

print("\n")
print("Column List: ")
yellow_tripdata_df_columns = yellow_tripdata_df.columns
print(yellow_tripdata_df_columns.tolist())

print("\n")
print("Data Types: ")
print(yellow_tripdata_df.dtypes)

print("\n")
print("Number of Nulls:")
yellow_tripdata_df_nulls = yellow_tripdata_df.isnull()
yellow_tripdata_df_sumofnulls = yellow_tripdata_df_nulls.sum()
yellow_tripdata_df_first10 = yellow_tripdata_df_sumofnulls.head(10)
print(yellow_tripdata_df_first10)
```

```python
#Taxi Zone Lookup Data
print("Number of Rows: ")
print(taxi_zone_lookup_df.shape[0])

print("\n")
print("Column List: ")
taxi_zone_lookup_df_columns = taxi_zone_lookup_df.columns
print(taxi_zone_lookup_df_columns.tolist())

print("\n")
print("Data Types: ")
print(taxi_zone_lookup_df.dtypes)

print("\n")
print("Number of Nulls:")
taxi_zone_lookup_df_nulls = taxi_zone_lookup_df.isnull()
taxi_zone_lookup_df_sumofnulls = taxi_zone_lookup_df_nulls.sum()
taxi_zone_lookup_df_first10 = taxi_zone_lookup_df_sumofnulls.head(10)
print(taxi_zone_lookup_df_first10)
```
